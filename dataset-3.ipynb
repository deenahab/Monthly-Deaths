### Filtering Further

For both of these subsets, the column *Year* including *2014* and *2019* will be explored.

As well as the columns *All Cause*, *Natural Cause*, *Septicemia*, and *Diseases of Heart*

These specific variables were chosen to explore the *All cause* and *Natural cause* numbers of heart and blood related diseases during the first and last year of this study.

Then a copy will be made to use it later on

Now create a name for this subset and display the subset

```python
filtered_subset = df[df["Year"] == 2014].copy()
```

```python
filtered_subset
```

Now we want to filter the subset even further by filtering by:


*  Month
*  All Cause
*  Natural Cause
*  Septicemia
*  Diseases of Heart

Using the following code and display the subset as well:

```python
filtered_subset=df.loc[(df["Year"] == 2014), ["Year","Month","All Cause","Natural Cause", "Septicemia","Diseases of Heart"]]
```

```python
filtered_subset
```

### Now creating the second subset

For this second subset, we will be filtering by the year "2019"

Firstly, create a name for this subset and display the subset

```python
filtered2_subset=df[df["Year"]==2019].copy()
```

```python
filtered2_subset
```

Now we want to filter the subset even further by filtering by:


*  Month
*  All Cause
*  Natural Cause
*  Septicemia
*  Diseases of Heart

Using the following code and display the subset as well:

```python
filtered2_subset=df.loc[(df["Year"] == 2019), ["Year","Month","All Cause","Natural Cause", "Septicemia","Diseases of Heart"]]
```

```python
filtered2_subset
```

Now we have filtered both of our new subsets

### Merging Data

Now we want to merge our two subsets into one by using the code **pd.concat()**   function

We want to put the 2014 subset before the 2019 so it will be in order

Display the new merged data set

```python
mergedData=pd.concat([filtered_subset,filtered2_subset]).copy()
```

```python
mergedData
```

To keep the dataset concise, the code should be added to the pd.concat() function


*   ignore_index=True

Then display the dataset once again, then it will be exported into a csv.

```python
mergedData=pd.concat([filtered_subset,filtered2_subset],ignore_index=True).copy()
```

```python
mergedData
```

### Exporting the new subset as a csv. file

To export the data into a csv. file, the following code will be used:

```python
mergedData.to_csv("2014_2019_Data.csv",index=False)
```
