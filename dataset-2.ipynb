### Concatenating Data

Used for putting together datasets that are same structurally.

data0 contains "All Cause" and "Natural Cause" and data1 contains "Cause_Change" which shows the difference beween "All Cause" and "Natural Cause".

```python
data0 = pd.DataFrame({"All Cause": [1, 2],
                    "Natural Cause": np.random.normal(0, 1, 2)}) #np.random.normal(0,1,2) pulls 2 random values from a Normal(0,1) distribution
data0
```

```python
data1=pd.DataFrame({"All Cause":[1,2],
                    "Cause_Change":np.random.normal(0,1,2),
                    "Month":["1", "2"]})
data1
```

### Merging Data

```python
pd.concat([data0,data1],axis=0,ignore_index=True,sort=False)
```

```python
series0 = pd.DataFrame({"Natural Cause": ["Septicemia", "Diseases of Heart"],
                      "Year": [2014, 2014]})
series1 = pd.DataFrame({"Natural Cause": ["Diseases of Heart", "Septicemia"],
                     "All Cause": [1, 2]})
```

```python
series0
```

```python
series1
```

```python
pd.merge(series0,series1,on="Natural Cause")
```
